## Disabling Microsoft OneDrive

Microsoft OneDrive is likely to be a major barrier for our work, unless you are paying for an account that allows for a lot of storage.

Due to this barrier, we would like you to disable it so that when we collect a large amount of data, it will not lock up your computer.

1. Find the OneDrive icon in the system tray on the far right of your toolbar.
      
    
1. Click on the OneDrive icon to open application and click on the settings (gear) icon in the top right.
      

1. Click on the `Pause Syncing` option and select `Quit OneDrive`.
      

1. Open the `Start Menu` and search for `Startup Apps` and open that application.
    
    
1. Find Microsoft OneDrive and turn it off.

## Forking and Cloning the Repository

GitHub is a great way to access existing code, while also being able to do your own work with that code without it being lost. It also has a built in history, so that changes you make can be stored along the way, think of it as a stored "undo" history. As long as you save your changes along the way (called commits), you can always get back to something that you did.

GitHub stores this code, the history, and the various versions (called branches) in what is called a **repository**.

Most of the code we will be working with already exists.

### Forking the Repository

In order for you to work with the repository, you will first need to create a 'fork' of it into your own account. Detailed instructions for creating a fork can be found [here](https://docs.github.com/en/pull-requests/collaborating-with-pull-requests/working-with-forks/fork-a-repo).

1. While logged into GitHub.com, open the URL above, or access it through the link shared in Discord.

1. Once you're on the website, click on the `Fork` button, usually near the top right of the page.
    
    
1. This action should create a version of the repository for your account. Verify that you can see the new repository by clicking on the cat icon in the top left and seeing the new repository for your account in the list on the left.
    

### Cloning the Repository onto your Computer

The fork you just created now exists for your account on GitHub.com, but it's not accessible on your computer.

You now want to create a `Clone` of that repository on your computer. Cloning a repository let's you make changes, but then choose when to upload those changes back to the original repository. This is useful because you can keep a "cloud" version of your code in the original repository, but get messy with trying out things on your computer. If you end up in a place where your code stops working, or gets too messy, you can go back to what is in the GitHub.com repository and throw away what you have on your computer.

1. Open GitHub Desktop

1. Click on the `File` menu option and select `Clone Repository`
    

1. A pop-up should open allowing you to select your forked version of the repository shown under "Your repositories" in the GitHub.com tab. Select the repository, but **DO NOT CLICK THE `Clone` BUTTON YET**.
    

1. Choose a `Local path` for this clone to be `C:\Users\YOUR_USER_NAME\Documents\GitHub`. Do not use a local path that is attached to OneDrive, it might have defaulted there. If you are unsure, ask for help. 

    For example, if your account username was `'ian'`, then the local path should read as below.
    


1. Once you have everything correct, click the `Clone` button at the bottom of the pop-up.
